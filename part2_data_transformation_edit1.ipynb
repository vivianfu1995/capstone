# Functions
Data transformation steps
1. Scale numerical variables
2. Reduce levels in categorical variables

```python
def scaler(df,col_name):
    X = np.array(df[col_name])
    X = preprocessing.scale(X)
    df[col_name] = X
    return df

# function:  reduce_level
# purpose:   some categorical variables have too mancol_name levels, we want to reduce levels 
# input:     method: a string, choose from --> 'qcut' or 'equal cut'
#                    'qcut': short for quantile cut - reduce levels based on quantiles 
#                            the bins will be chosen so that col_nameou have the same number of records in each bin.            
#                    'number': reduce levels based on equal intervals
#                            example: '2','3','4'.... a number in str
def reduce_level(df,col_name,method):
    if method == 'qcut':
        df[col_name] = pd.qcut(df[col_name], 4,labels=False)
    else: #  method is some number in str
        interval = int(method) 
        df[col_name] = pd.cut(df[col_name], interval,labels=False)
    return df
            
```

## 2. Reduce levels in categorical variables
The following features from the dataframe are categorical and have too many levels:
- family income
- education
- marital status

```python
# Before reducing 
df[['family income','education','marital status']].head()
```

```python
df = reduce_level(df,'family income','qcut')
df = reduce_level(df,'education','2')
```

To reduce levels in marital status, needs to be more customized. 

    The original codes: 
    1	Married		
    2	Widowed
    3	Divorced	
    4	Separated	
    5	Never married		
    6	Living with partner	
    
    New codes:
    1   Married & Living with partner	
    2   Widowed & Divorced & Separated
    3   Never married

```python
df['marital status'][df['marital status']==6]=1
df['marital status'][df['marital status']==3]=2
df['marital status'][df['marital status']==4]=2
df['marital status'][df['marital status']==5]=3
```
